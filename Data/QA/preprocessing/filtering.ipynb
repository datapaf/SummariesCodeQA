# Train

```python
import json

data_dir = "../UltimateQADataset_filtered/UltimateQADataset_filtered/"

with open(data_dir + 'filtered_train.py.jsonl', 'r') as json_file:
    train_data = [json.loads(line) for line in json_file]
```

```python
idxs = []
keywords = [
    ':param',
    '>>>',
    'args:',
    'returns:',
    'cli example:',
    '@param',
    ':rtype',
    ':type',
    '@type',
    'http://',
    'https://',
    'see also:',
    ':return:',
    '    ',
    # 'return'
]

for i, ex in enumerate(train_data):
    answer = ex['answers'] if 'answers' in ex else ex['answer']
    for keyword in keywords:
        if keyword in answer.lower():
            idxs.append(i)
            print(answer)
            print()
            break
```

```python
len(idxs)
```

```python
import re

def check_string(string):
    pattern = r'^[a-zA-Z0-9\s!"#$%&\'()*+,-./:;<=>?@\[\\\]^_`{|}~]+$'
    result = re.match(pattern, string)
    if result:
        return True
    else:
        return False
```

```python
#idxs = []

for i, ex in enumerate(train_data):
    if not check_string(ex['code']):
        idxs.append(i)
        print(ex['code'])
        print('><')
```

```python
idxs = list(set(idxs))
```

```python
len(idxs)
```

```python
train_data_filtered = []

for i, ex in enumerate(train_data):
    if i in idxs:
        continue
    train_data_filtered.append(ex)
```

```python
len(train_data), len(train_data_filtered)
```

```python
import jsonlines

with jsonlines.open('../UltimateQADataset/train_filtered2.py.jsonl', mode='w') as writer:
    writer.write_all(train_data_filtered)
```
