## Test

```python
import json
with open('../UltimateQADataset/test_filtered2.py.jsonl') as f:
    test_data = [json.loads(line) for line in f]
```

```python
test_data[-500]
```

```python
print(test_data[228]['code'])
```

```python
cs1qa = [item for item in test_data if item['src'] == 'cs1qa']
codeqa = [item for item in test_data if item['src'] == 'codeqa']
```

```python
len(cs1qa), len(codeqa)
```

```python
import pandas as pd

def compose_df(test_data):
    return pd.DataFrame({
    "question": [x['question'].strip()  for x in test_data],
    "code": [x['code']  for x in test_data],
    "answer": [x["answers"].strip() if x['src'] == 'codeqa' else x["answer"].strip() for x in test_data]
})
```

```python
df = compose_df(test_data)
df = df.sample(frac=1).reset_index(drop=True)
```

```python
from fastparquet import write
write('../UltimateQADataset/test_filtered2.parquet', df)
```

CS1QA

```python
cs1qa_df = compose_df(cs1qa)
cs1qa_df
```

```python
from fastparquet import write
write('../UltimateQADataset/cs1qa_filtered2_test.parquet', cs1qa_df)
```

CodeQA

```python
codeqa_df = compose_df(codeqa)
codeqa_df
```

```python
from fastparquet import write
write('../UltimateQADataset/codeqa_filtered2_test.parquet', codeqa_df)
```
